```python
import numpy as np
import matplotlib.pyplot as plt

from matplotlib import animation, rc
from IPython.display import HTML
import random
```

```python
def init():
    point.set_data([],[])
    point_start.set_data([],[])
    point_end.set_data([],[])
    line.set_data([], [])
    line1.set_data([], [])
    line2.set_data([], [])
    line3.set_data([], [])
    line4.set_data([], [])
    return (line,line1,line2,line3,line4,point,point_start,point_end)

# animation function.  This is called sequentially

def animate(i_frame):
    n_points = 1000
    line_cut_1 = int(round((1/4)*n_points))
    line_cut_2 = int(round((2/4)*n_points))
    line_cut_3 = int(round((3/4)*n_points))
    i_point = int(round((5/8)*n_points))
    start = 0
    stop = 1
    x = np.linspace(start,stop,n_points)
    f = 3
    phase_step = 0.01
    y = np.sin(2*np.pi*f*(x-phase_step*i_frame))
    line.set_data(x,y)
    line1.set_data(x[0:line_cut_1], y[0:line_cut_1])
    line2.set_data(x[line_cut_1:line_cut_2], y[line_cut_1:line_cut_2])
    line3.set_data(x[line_cut_2:line_cut_3], y[line_cut_2:line_cut_3])
    line4.set_data(x[line_cut_3:n_points-1], y[line_cut_3:n_points-1])
    point.set_data(x[i_point],y[i_point])
    point_start.set_data(x[0],y[0])
    point_end.set_data(x[-1],y[-1])
    return line,line1,line2,line3,line4,point,point_start,point_end
```

```python
fig = plt.figure(figsize=(12,6))
ax = plt.axes(xlim=(-0.2, 1.2), ylim=(-2, 2))
ax.axis('off')
line, = ax.plot([], [], color='#000000', linewidth = 10)
line1, = ax.plot([], [], color='#ff0000', linewidth = 5)
line2, = ax.plot([], [], color='#00ff00', linewidth = 5)
line3, = ax.plot([], [], color='#0000ff', linewidth = 5)
line4, = ax.plot([], [], color='#ffff00', linewidth = 5)
point_start, = ax.plot([], [],'o',color='k',markersize=15)
point_end, = ax.plot([], [],'o',color='k',markersize=15)
point, = ax.plot([], [],'o',color='r',markersize=25)
ax.arrow(0.5, 1.5, 0.1, 0, width=0.15,head_width=0.4, head_length=0.05, fc='k', ec='k');
anim = animation.FuncAnimation(fig, animate, init_func=init, frames=100, interval=120, blit=True);
HTML(anim.to_html5_video())
```

```python
def init():
    ball.set_data([],[])
    point_along.set_data([],[])
    spring.set_data([],[])
    return (ball,point_along,spring)

# animation function.  This is called sequentially
def animate(i_frame):
    n_points = 1000
    start = 0
    stop = 1
    x = np.linspace(start,stop,n_points)
    f = 3
    phase_step = 0.01
    y = np.sin(2*np.pi*f*(x+phase_step*i_frame))
    point_along.set_data(x[i_frame]+phase_step*i_frame,y[i_frame])
    ball.set_data(-0.1,y[i_frame])
    
    freq_mult=0.9
    
    spring_y=np.linspace(1,y[i_frame])
    spring_x=-0.04 + 0.05*np.cos(freq_mult*spring_y*(20 - 10*np.abs(1-y[i_frame]))-np.pi/2) - 0.05*np.cos(freq_mult*spring_y*(20 - 10*np.abs(1-y[i_frame])))
    
    spring.set_data(spring_x,spring_y)
    
    return ball,point_along,spring
```

```python
fig,axs = plt.subplots(1,2,figsize=(12,6),sharey=True,gridspec_kw={'width_ratios': [1, 5]})
axs[1].set_xlim([0,0.99])
axs[1].set_ylim([-1.5,1.5])
axs[0].set_xlim([-1,0.1])
axs[0].axis('off')
axs[1].set_xlabel('Time')
axs[1].set_ylabel('Height')
axs[1].set_xticks([])
axs[1].set_yticks([])
axs[1].set_title('Height of the ball over time')
n_points = 1000
start = 0
stop = 1
f = 3
x = np.linspace(start,stop,n_points)
y = np.sin(2*np.pi*f*x)
spring, = axs[0].plot([], [],linestyle='-',color='#999999',linewidth=2)
line, = axs[1].plot(x, y, color='#666666', linewidth = 3)
ball, = axs[0].plot([], [],'o',color='r',markersize=35)
point_along, = axs[1].plot([], [],'o',color='k',markersize=15)
anim = animation.FuncAnimation(fig, animate, init_func=init, frames=100, interval=70, blit=True)
HTML(anim.to_html5_video())
```

```python
def init():
    line.set_data([],[])
    line1.set_data([],[])
    line2.set_data([],[])
    point_start.set_data([],[])
    point_stop.set_data([],[])
    return (line,line1,line2,point_start,point_stop)

# animation function.  This is called sequentially
def animate(i_frame):
    n_points = 1000
    start = 0
    stop = 1
    x = np.linspace(start,stop,n_points)
    frequency = 4
    phase_step = 0.0025
    y1 = 0.5*np.sin(2*np.pi*frequency*(x+phase_step*i_frame))
    y2 = 0.5*np.sin(2*np.pi*frequency*(x-phase_step*i_frame))
    y = y1 + y2
    line1.set_data(x,y1)
    line2.set_data(x,y2)
    line.set_data(x,y)
    point_start.set_data(x[0],y[0])
    point_stop.set_data(x[-1],y[-1])
    return line, line1,line2, point_start, point_stop

fig,ax = plt.subplots(1,1,figsize=(12,6))

ax.set_ylim([-1.5,1.5])
ax.set_xlim([-0.5,1.5])
ax.axis('off')


line1, = ax.plot([], [],linestyle='--',color='#ff0000',linewidth=3)
line2, = ax.plot([], [], linestyle='--',color='#0000ff', linewidth = 3)
line, = ax.plot([], [],linestyle='-',color='#ff00ff',linewidth=4)
point_start, = ax.plot([], [],'o',color='k',markersize=15)
point_stop, = ax.plot([], [],'o',color='k',markersize=15)
anim = animation.FuncAnimation(fig, animate, init_func=init, frames=300, interval=40, blit=True)
HTML(anim.to_html5_video())
```

```python
def add_note(x,y,freq,amplitude,color='k',linewidth=2,zorder=1):
    sigma = x[-1]-x[0])/x.shape[0]
    y += amplitude * np.exp( -((x-freq) / sigma)**2)
    return x,y
```

```python

n_points = 10000
n_harmonics = 10
max_amplitude = 1
min_amplitude = 0.1
freq = 440
colors = ['r','b','g']
#n_freqs = len(freqs)
instruments = ['Violin','Flute','Piano']
n_instruments = len(instruments)
amplitudes = np.zeros([n_harmonics,n_instruments])
amplitudes[:,0] = np.array([1 , 0.55 , 0.58 , 0.7 , 0.58 , 0.4 , 0.55 , 0.55 , 0.35 , 0.15]) # Violin
amplitudes[:,1] = np.array([1 , 0.9 , 0.1 , 0.2 , 0.15 , 0.05 , 0.05 , 0.05 , 0.05 , 0.05])
amplitudes[:,2] = np.array([1 , 0.6 , 0.15 , 0.25 , 0.05 , 0.5 , 0.05 , 0.1 , 0.35 , 0.25]) # Piano

stop = freq*n_harmonics*1.1
start = 0

x = np.linspace(start,stop,n_points)
y = np.zeros((n_instruments,n_points))

sigma = (x[-1]-x[0])/x.shape[0]*10

fig, ax = plt.subplots(n_instruments,1,figsize=[14,(n_instruments+1)*5])
plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=None, hspace=None)


for i_inst in range(n_instruments):
    for i_harmonic in range(n_harmonics):
        y[i_inst,:] += amplitudes[i_harmonic,i_inst] * np.exp( -((x-freq * (i_harmonic+1)) / sigma)**2)
        ax[i_inst].plot(freq * (i_harmonic+1),amplitudes[i_harmonic,i_inst],'o',markersize=10,zorder=1,color=colors[i_inst])
    
    ax[i_inst].plot(x,y[i_inst,:],colors[i_inst],2,1)
    ax[i_inst].set_title(instruments[i_inst])
    ax[i_inst].set_xlim([start,stop])
    ax[i_inst].set_ylim([-0.1,1.1])
    ax[i_inst].set_xlabel("Frequency (Hz)")
    ax[i_inst].set_ylabel("Amplitude")
        
```

```python



n_points = 10000
n_harmonics = 20
max_amplitude = 1
min_amplitude = 0.1
freqs = np.zeros(2)
frequency_ratio_num = 7
frequency_ratio_denom = 3
frequency_ratio = frequency_ratio_num/frequency_ratio_denom
freqs[0] = 100
freqs[1] = freqs[0] * (frequency_ratio)
colors = ['orange','b']
n_freqs = len(freqs)
amplitudes = np.zeros([n_harmonics,n_freqs])

stop = max(freqs)*n_harmonics*1.1
start = 0

fig, ax = plt.subplots(n_freqs+1,1,figsize=[14,(n_freqs+1)*5])
plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=None, hspace=None)

x = np.linspace(start,stop,n_points)
y = np.zeros((n_freqs,n_points))

sigma = (x[-1]-x[0])/x.shape[0]*10

for i_freq in range(n_freqs+1):
    if i_freq == 0:
        ax[i_freq].text(stop*(43/100), 0.8, 'Frequency ratio =', fontsize = 36, color = 'k')
        ax[i_freq].text(stop*(35/40), 0.85, str(frequency_ratio_num), fontsize = 36, color = colors[1])
        ax[i_freq].text(stop*(347/400), 0.85, '__', fontsize = 36, color = 'k')
        ax[i_freq].text(stop*(35/40), 0.66, str(frequency_ratio_denom), fontsize = 36, color = colors[0])
    if i_freq<n_freqs:
        for i_harmonic in range(n_harmonics):
            amplitudes[i_harmonic,i_freq] = max_amplitude-i_harmonic*(max_amplitude-min_amplitude)/n_harmonics
            y[i_freq,:] += amplitudes[i_harmonic,i_freq] * np.exp( -((x-freqs[i_freq] * (i_harmonic+1)) / sigma)**2)
            ax[i_freq].plot(freqs[i_freq] * (i_harmonic+1),amplitudes[i_harmonic,i_freq],'o',markersize=10,zorder=1,color=colors[i_freq])
            ax[-1].plot(freqs[i_freq] * (i_harmonic+1),amplitudes[i_harmonic,i_freq],'o',markersize=10,zorder=1,color=colors[i_freq])
    
        ax[i_freq].set_title("Fundamental Frequency = "+str(freqs[i_freq])+" Hz")
        ax[i_freq].plot(x,y[i_freq,:],colors[i_freq],2,1)
        ax[-1].plot(x,y[i_freq,:],colors[i_freq],2,1)
    
    ax[i_freq].set_xlim([start,stop])
    ax[i_freq].set_xlabel("Frequency (Hz)")
    ax[i_freq].set_ylabel("Amplitude")
```
