# Wave travelling on a string

A wave can also represent displacements along a medium. In the animation below, the wave travels from left to right, but the string only moves up and down. The string is the **medium**, and we say that the wave is a **transverse wave** because the direction of travel of the wave is perpendicular to the direction of the displacement of the medium.

<center>
<video src="4_colors_string_no_ball.mp4" controls loop autoplay></video>
</center>

# Standing wave

When two waves of the same frequency and amplitude are travelling in opposite direction in the same medium, it creates a standing wave. In the animation below, the red wave is travelling to the left, and a blue wave is travelling to the right. The pink wave is the resulting standing wave.
