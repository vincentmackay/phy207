```python
def draw_function():
    n_points = 10000
    start = 0
    stop = 1
    width = 800  # canvas width
    height = 200 # canvas height
    center = height//2
    white = (255, 255, 255) # canvas back

    def save():
        # save image to hard drive
        filename = "user_input.png"
        output_image.save(filename)
        master.destroy()

    def paint(event):
        polygon_half_width = 4
        x1, y1 = (event.x - 1), (event.y - 1)
        x2, y2 = (event.x + 1), (event.y + 1)
        #canvas.create_oval(x1, y1, x2, y2, fill="black",width=5)
        canvas.create_line(x1+10, y1-5, x1+10, 0, fill="white",width=10)
        canvas.create_line(x1+10, y1+5, x1+10, height, fill="white",width=10)
        draw.line([x1+10, y1-5, x1+10, 0], fill="white",width=10)
        draw.line([x1+10, y1+5, x1+10, height], fill="white",width=10)
        canvas.create_polygon( x1+polygon_half_width, y1+polygon_half_width, x1+polygon_half_width, y1-polygon_half_width, x1-polygon_half_width, y1-polygon_half_width, x1-polygon_half_width, y1+polygon_half_width, fill="black",width=5)
        draw.rectangle( [x1+polygon_half_width, y1+polygon_half_width, x1-polygon_half_width, y1-polygon_half_width], fill="black")


    master = Tk()

    # create a tkinter canvas to draw on
    canvas = Canvas(master, width=width, height=height, bg='white')
    canvas.pack()

    # create an empty PIL image and draw object to draw on
    output_image = PIL.Image.new("RGB", (width, height), white)
    draw = ImageDraw.Draw(output_image)
    canvas.pack(expand=YES, fill=BOTH)
    canvas.bind("<B1-Motion>", paint)

    # add a button to save the image
    button=Button(text="save",command=save)
    button.pack()

    master.mainloop()

    im = Image.open('user_input.png')
    im = im.convert('1')
    width, height = im.size
    pix = np.array(np.array(im.getdata())).reshape((height, width))

    n_points_drawing = 0

    for i in range(width):
        if np.min(pix[:,i])==0:
            if n_points_drawing==0:
                first_point = i
            n_points_drawing += 1
            last_point = i

    start_drawing = 0
    stop_drawing = 1
    x_drawing = np.linspace(start_drawing , stop_drawing , n_points_drawing)
    y_drawing = np.zeros( n_points_drawing )
    i_x = 0
    for i in range(width):
        if np.min(pix[:,i])==0:
            x_drawing[i_x] = i
            y_drawing[i_x] = height-np.argmin(pix[:,i])
            i_x+=1
    x_drawing = (x_drawing - min(x_drawing) )/max(x_drawing)
    y_drawing =  10*y_drawing/max(y_drawing)
    return x_drawing, y_drawing
    
def make_complex_tone(n_funcs,pitched=True):

    start = 0
    stop = 1
    n_points = 1000
    freq = 2

    x = np.linspace(start , stop , n_points)
    y = np.zeros([n_points,n_funcs])
    fundamental_frequency = 10
    freqs = np.zeros(n_funcs)
    amplitudes = np.zeros(n_funcs)
    for i in range(n_funcs):
        if i==0:
            freqs[i] = fundamental_frequency
            amplitudes[i] = 1*pitched + (0.4 + 0.6 * random.random())*(1-pitched)
        else:
            freqs[i] =  fundamental_frequency * (i + 1 + random.random()*(1-pitched))
            amplitudes[i] = (0.4 + 0.6 * random.random())
        y[:,i] = amplitudes[i] * np.sin(2 * np.pi * freqs[i] * (x+random.random()))


    ylim = 1.2 * np.max(np.abs(np.sum(y,1)))
    rect_opacity = 0.1

    for i in range(n_funcs):

        plt.figure(figsize=[14,2])
        gs = gridspec.GridSpec(1, 2, width_ratios=[5, 1])
        period = 1/fundamental_frequency
        ax0 = plt.subplot(gs[0])
        ax1 = plt.subplot(gs[1])
        ax0.plot(x,y[:,i])
        for j in range(fundamental_frequency+1):
            if j%2==0:
                rect = Rectangle([j*period,-ylim],period,2*ylim,alpha=rect_opacity,color='k')
                ax0.add_patch(rect)
        ax0.set_xlabel('Time (s)')
        ax0.set_ylabel('Pressure change')
        ax0.set_xlim([0,1])
        ax0.set_ylim([-ylim,ylim])
        ax0.set_title('Pure tone, frequency = '+ "{:.2f}".format(freqs[i]) +' Hz')
        ax1.plot(freqs[i],amplitudes[i],'.',markersize=20)
        
        n_points_xy = 1000
        xy_f = np.zeros([2,n_points_xy])
        xy_f[0,:] = np.linspace(0,freqs[-1]+10,n_points_xy)
        for j in range(n_points_xy)[1:]:
            if xy_f[0,j-1]<freqs[i] and xy_f[0,j]>=freqs[i]:
                xy_f[1,j] = amplitudes[i]
                break


        ax1.plot(xy_f[0,:],xy_f[1,:], linestyle='-',color='k',alpha=1)
        
        
        
        
        ax1.set_xlim([0,freqs[-1]+10])
        ax1.set_ylim([0,np.max(amplitudes)*1.1])
        ax1.set_xlabel('Frequency (Hz)')
        ax1.set_ylabel('Amplitude')
        ax1.set_title( '\"Fourier representation\"')
    
    if n_funcs > 1:
        plt.figure(figsize=[14,2])
        gs = gridspec.GridSpec(1, 2, width_ratios=[5, 1])
        ax0 = plt.subplot(gs[0])
        ax1 = plt.subplot(gs[1])
        for i in range(n_funcs):
            ax0.plot(x,y[:,i],alpha=1)   
        for j in range(fundamental_frequency+1):
            if j%2==0:
                rect = Rectangle([j*period,-ylim],period,2*ylim,alpha=rect_opacity,color='k')
                ax0.add_patch(rect)
        ax0.set_xlim([0,1])
        ax0.set_ylim([-ylim,ylim])
        ax0.plot(x,np.sum(y,1),linewidth=3,color='k',linestyle='-')
        ax0.set_xlabel('Time (s)')
        ax0.set_ylabel('Pressure change')
        ax0.set_title('Complex tone, sum of all pure tones above')
        for i in range(n_funcs):
            ax1.plot(freqs[i],amplitudes[i],'.',markersize=20)
            
        n_points_xy = 1000
        xy_f = np.zeros([2,n_points_xy])
        xy_f[0,:] = np.linspace(0,freqs[-1]+fundamental_frequency,n_points_xy)
        for i in range(n_funcs):
            for j in range(n_points_xy)[1:]:
                if xy_f[0,j-1]<freqs[i] and xy_f[0,j]>=freqs[i]:
                    xy_f[1,j] = amplitudes[i]
                    break

        ax1.plot(xy_f[0,:],xy_f[1,:], linestyle='-',color='k',alpha=1)
            
            
        #ax1.plot( np.real( freqs[0:n_funcs] ) , amplitudes[0:n_funcs], linestyle='-',color='k',alpha=1)
        ax1.set_xlim([0,freqs[-1]+10])
        ax1.set_ylim([0,np.max(amplitudes)*1.1])
        ax1.set_xlabel('Frequency (Hz)')
        ax1.set_ylabel('Amplitude')
        ax1.set_title( '\"Fourier representation\"')


def make_fourier_plots(x,y,n_freqs_to_plot,one_by_one=False):


    xf = np.fft.fftfreq(n_points,(stop - start) / n_points)
    yf = np.fft.fft( y )

    plt.figure(figsize=[14,2])
    gs = gridspec.GridSpec(1, 2, width_ratios=[5, 1])
    ax0 = plt.subplot(gs[0])
    ax1 = plt.subplot(gs[1])

    ax0.plot(x,y - np.real(yf[0])* (1 / len(x) ),linewidth =5,color = 'k')
    ax0.set_title('Original complex tone')
    ax0.set_xlabel('Time (s)')
    ax0.set_ylabel('Pressure change')
    ax0.set_xlim([0,1])

    to_plot = np.zeros(n_points)

    threshold = 0

    ax1.set_xlim([0,1])
    ax1.set_ylim([0,1])
    ax1.text(0.5,0.5,'?',ha='center',va='center',fontsize=50)
    ax1.set_xlabel( 'Frequency (Hz)' )
    ax1.set_ylabel( 'Amplitude' )
    ax1.set_title( '\"Fourier representation\"')

    plt.figure(figsize=[14,2])
    gs = gridspec.GridSpec(1, 2, width_ratios=[5, 1])
    ax0 = plt.subplot(gs[0])
    ax1 = plt.subplot(gs[1])
    
    for i in range(n_freqs_to_plot+1)[1:]:
        if abs(yf[i])/(len(x)) > threshold:
            to_add = (1 / len(x) ) * ( np.real(yf[i]) * np.cos( 2 * np.pi * xf[i] * x ) - np.imag(yf[i]) * np.sin( 2 * np.pi * xf[i] * x ) )
            to_plot += to_add
            ax0.plot (x , to_add  , linestyle = '-',linewidth=2)#float(i)/n_freqs_to_plot/2)
            
                        
            
            
            
            

    ax0.plot( x , to_plot,linewidth =5,color = 'k')
    ax0.set_title('Original complex tone as a sum of ' + str(n_freqs_to_plot) +' pure tone(s)')
    ax0.set_xlabel('Time (s)')
    ax0.set_ylabel('Pressure change')
    ax0.set_xlim([0,1])

    for i in range(n_freqs_to_plot+1)[1:]:
        ax1.plot( np.real( xf[i] ) , np.abs( yf[i] * (1 / len(x) ) ), '.' ,markersize=20)
        ax1.set_xticks(np.arange(n_freqs_to_plot)+1)
        
        
        
        
        
        
        
        
        
    
    
    n_points_xy = 1000
    xy_f = np.zeros([2,n_points_xy])
    xy_f[0,:] = np.linspace(0,np.real( xf[i] )+1,n_points_xy)
    for i in range(n_freqs_to_plot+1):
        for j in range(n_points_xy)[1:]:
            if xy_f[0,j-1]<np.real( xf[i] ) and xy_f[0,j]>=np.real( xf[i] ):
                xy_f[1,j] = np.abs( yf[i] * (1 / len(x) ) )
                break
    
    #if n_freqs_to_plot>1:
        #ax1.plot( np.real( xf[1:n_freqs_to_plot+1] ) , np.abs( yf[1:n_freqs_to_plot+1] * (1 / len(x) ) ), linestyle='-',color='k',alpha=1)
    ax1.plot(xy_f[0,:],xy_f[1,:], linestyle='-',color='k',alpha=1)
                         
    ax1.set_xlabel( 'Frequency (Hz)' )
    ax1.set_ylabel( 'Amplitude' )
    ax1.set_title( '\"Fourier representation\"')
    ax1.set_xlim([0,xf[n_freqs_to_plot+1]])
    ax1.set_ylim([0,1.1*np.max( np.abs( yf[1:] * (1 / len(x) ) ) )])#(1 / len(x) ) * np.max(np.abs(yf[1:n_freqs_to_plot+1]))])
    
    if one_by_one:

        for j in range(n_freqs_to_plot+1)[1:]:

            plt.figure(figsize=[14,2])
            gs = gridspec.GridSpec(1, 2, width_ratios=[5, 1])
            ax0 = plt.subplot(gs[0])
            ax1 = plt.subplot(gs[1])
            to_plot2 = np.zeros(n_points)
            #plt.subplot(n_freqs_to_plot,1,j)
            for i in range(j+1)[1:]:
                if abs(yf[i])/(len(x)) > threshold:
                    to_add = (1 / len(x) ) * ( np.real(yf[i]) * np.cos( 2 * np.pi * xf[i] * x ) - np.imag(yf[i]) * np.sin( 2 * np.pi * xf[i] * x ) )
                    to_plot2 += to_add
                    if i<j:
                        ax0.plot (x , to_add  , linestyle = '-',linewidth=2,alpha=0.25)#float(i)/n_freqs_to_plot/2)
                        ax1.plot( np.real( xf[i] ) , np.abs( yf[i] * (1 / len(x) ) ), '.' ,markersize=20,alpha=0.25)
                    else:
                        ax0.plot (x , to_add  , linestyle = '-',linewidth=2,alpha=1)#float(i)/n_freqs_to_plot/2)
                        ax1.plot( np.real( xf[i] ) , np.abs( yf[i] * (1 / len(x) ) ), '.' ,markersize=20,alpha=1)
                ax1.set_xticks(np.arange(n_freqs_to_plot)+1)
                
                n_points_xy = 1000
                xy_f = np.zeros([2,n_points_xy])
                xy_f[0,:] = np.linspace(0,np.real( xf[n_freqs_to_plot] )+1,n_points_xy)
                for k in range(i+1):
                    for l in range(n_points_xy)[1:]:
                        if xy_f[0,l-1]<np.real( xf[k] ) and xy_f[0,l]>=np.real( xf[k] ):
                            xy_f[1,l] = np.abs( yf[k] * (1 / len(x) ) )
                            break

                #if i>1:
                    #ax1.plot( np.real( xf[1:i+1] ) , np.abs( yf[1:i+1] * (1 / len(x) ) ), linestyle='-',color='k',alpha=1)
                ax1.plot(xy_f[0,:],xy_f[1,:], linestyle='-',color='k',alpha=1)
                
            ax0.plot( x , to_plot2,linewidth =5,color = 'k')
            ax0.set_ylabel('Pressure change')
            ax0.set_xlabel('Time (s)')
            ax0.set_xlim([0,1])
            ax1.set_xlabel( 'Frequency (Hz)' )
            ax1.set_ylabel( 'Amplitude' )
            ax1.set_title( '\"Fourier representation\"')
            ax1.set_xlim([0,xf[n_freqs_to_plot+1]])
            ax1.set_ylim([0,1.1*np.max( np.abs( yf[1:] * (1 / len(x) ) ) )])#(1 / len(x) ) * np.max(np.abs(yf[1:n_freqs_to_plot+1]))])
        
```
